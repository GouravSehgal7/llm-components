let we have single layer neuron with 5 input

```python
import torch


class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        
        # Initialize weights manually using torch
        self.W1 = torch.randn(input_size, hidden_size)
        self.b1 = torch.zeros(hidden_size)

        self.W2 = torch.randn(hidden_size, output_size)
        self.b2 = torch.zeros(output_size)

    def forward(self, x):
        z1 = x @ self.W1 + self.b1
        z2 = z1 @ self.W2 + self.b2
        return z2
    
net = SimpleNN(input_size=3, hidden_size=4, output_size=2)
x = torch.tensor([1.2, -0.7, 3.0])
output = net.forward(x)
print("Output:", output)
```
